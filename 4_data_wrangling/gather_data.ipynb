```python
movie_list = []
for filename in os.listdir('rt_html'):
    if filename.endswith('.html'):
        html_file = os.path.join('rt_html', filename)
#         print (html_file)
        with open(html_file) as file:
            soup = bs(file, 'lxml')
            movie = {}
            movie['name'] = soup.find('h1', attrs = {'id':'movie-title'}).text
            movie['audience_score'] = soup.find('div', attrs = {'class':'audience-score'}).find('span').text
            movie['number_user_ratings'] = soup.find('div', attrs = {'class':'audience-info'}).findAll('div')[1].contents[2]
            movie_list.append(movie)
```
